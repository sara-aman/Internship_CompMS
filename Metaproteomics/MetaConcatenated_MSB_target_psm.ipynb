```python
# Define a function to extract text between the first and last dots
def extract_text(text):
    dot_index_1 = text.find('.')
    dot_index_2 = text.rfind('.')

    if dot_index_1 != -1 and dot_index_2 != -1 and dot_index_1 != dot_index_2:
        return text[dot_index_1 + 1:dot_index_2]
    else:
        return ""

# Apply the function to create a new column
MSB_filtered_psm_IGC ["modi_peptide"] = MSB_filtered_psm_IGC ['peptide'].apply(extract_text)

# Display the updated DataFrame
MSB_filtered_psm_IGC.tail(50)
```
