### improt packages

```python
import pandas
import re
```

## A. target peptides

```python
rescore_peptides_A1101 = pandas.read_csv("/home/thinkpad/HLA_Files/A1101/A1101_Oktoberfest_MSFragger/results/percolator/rescore_target.peptides", sep= '\t')
```

```python
rescore_peptides_A1101
```

```python
print (rescore_peptides_A1101.PSMId.unique ,'\n \n') 
print (rescore_peptides_A1101.peptide.unique ,'\n \n')
print (rescore_peptides_A1101.proteinIds.unique)
```

### filter the dataset with FDR = 0.01

```python
filtered_peptide_A1101 = rescore_peptides_A1101[rescore_peptides_A1101['q-value'] <= 0.01]

# modify the peptide in a new column
filtered_peptide_A1101["modi_peptide"] = filtered_peptide_A1101['peptide']

filtered_peptide_A1101
```

```python
filtered_peptide_A1101['modi_peptide'] = filtered_peptide_A1101['modi_peptide'].str.replace('[_,.,:,4,35]', '', regex= True)
filtered_peptide_A1101['modi_peptide'] = filtered_peptide_A1101['modi_peptide'].str.replace(r'\[UNIMOD\]', '', regex= True)

filtered_peptide_A1101.tail(20)
```

```python
filtered_peptide_A1101.to_csv('/home/thinkpad/HLA_Files/OK_MSfragger_filtered/filtered_target_peptides_A1101.csv')
```
