#### Filter to FDR <= 0.01 

```python
MSB_filtered_psm_A1101 = A1101_psm[A1101_psm['q-value'] <= 0.01]
MSB_filtered_psm_A1101
```

### Keep the peptide with proper format, keep only what's between first and last dots without the brackets content 

```python
# Define a function to extract text between the first and last dots
def extract_text(text):
    dot_index_1 = text.find('.')
    dot_index_2 = text.rfind('.')

    if dot_index_1 != -1 and dot_index_2 != -1 and dot_index_1 != dot_index_2:
        return text[dot_index_1 + 1:dot_index_2]
    else:
        return ""

# Apply the function to create a new column
MSB_filtered_psm_A1101 ["modi_peptide"] = MSB_filtered_psm_A1101 ['peptide'].apply(extract_text)

# Display the updated DataFrame
MSB_filtered_psm_A1101.tail(50)
```

```python
# Define a function to remove square brackets and their content
def remove_square_brackets(text):
    cleaned_text = re.sub(r'\[.*?\]', '', text)
    return cleaned_text

# Apply the function to create a new column
MSB_filtered_psm_A1101 ["modi_peptide_2"] = MSB_filtered_psm_A1101 ['modi_peptide'].apply(remove_square_brackets)

# Display the updated DataFrame
MSB_filtered_psm_A1101.tail(50)
```
